```python
# üìö BIBLIOTECAS E CONFIGURA√á√ïES GERAIS
# Projeto: Dataflow Sentinel ‚Äì MVP de Arquitetura de Dados
# Bloco 1: Setup Inicial e Ingest√£o de Dados


# Manipula√ß√£o de dados
import pandas as pd
import numpy as np

# Visualiza√ß√£o
import matplotlib.pyplot as plt
import seaborn as sns

# Estilo de plots
sns.set(style="whitegrid", palette="muted")
plt.rcParams["figure.figsize"] = (12, 6)

# Sistema e logging
import os
import logging

# Configurar logging
logging.basicConfig(level=logging.INFO,
                    format='[%(levelname)s] %(asctime)s - %(message)s',
                    datefmt='%H:%M:%S')
logger = logging.getLogger()
```

‚úÖ C√âLULA 2 ‚Äì Caminho e Leitura do CSV

#### üìÅ LEITURA DO ARQUIVO CSV SIMULADO

```python
# =============================================================
# üìÅ LEITURA DO ARQUIVO CSV SIMULADO
# =============================================================

# Definir o caminho do arquivo CSV carregado no Colab
csv_path = '/content/dataflow_sentinel_supplyops.csv'

# Verificar se o arquivo existe
if not os.path.exists(csv_path):
    logger.error(f"Arquivo CSV n√£o encontrado: {csv_path}")
else:
    logger.info(f"Arquivo localizado: {csv_path}")

# Carregar os dados
df = pd.read_csv(csv_path)

# Exibir as primeiras linhas
df.head()
```

‚úÖ C√âLULA 3 ‚Äì Preview do Dataset

#### üîç AN√ÅLISE INICIAL DOS DADOS

```python
# =============================================================
# üîç AN√ÅLISE INICIAL DOS DADOS
# =============================================================

# Tamanho do dataset
logger.info(f"Total de registros: {df.shape[0]}")
logger.info(f"Total de colunas: {df.shape[1]}")

# Tipos de dados
print("üìå Tipos de dados:")
display(df.dtypes)

# Valores ausentes
print("üßº Valores ausentes por coluna:")
display(df.isnull().sum())

# Estat√≠sticas b√°sicas
print("üìä Estat√≠sticas descritivas:")
display(df.describe())
```

‚úÖ C√âLULA 4 ‚Äì An√°lise Explorat√≥ria R√°pida

#### üìä AN√ÅLISE EXPLORAT√ìRIA VISUAL INICIAL

```python
# =============================================================
# üìä AN√ÅLISE EXPLORAT√ìRIA VISUAL INICIAL
# =============================================================

# Distribui√ß√£o de status das entregas
sns.countplot(data=df, x="status")
plt.title("Distribui√ß√£o de Status das Entregas")
plt.show()

# Boxplot de custos por rota
sns.boxplot(data=df, x="route_id", y="cost_usd")
plt.title("Custos por Rota")
plt.show()

# Temperatura m√©dia por status
sns.barplot(data=df, x="status", y="temp_c", estimator=np.mean)
plt.title("Temperatura M√©dia por Status")
plt.show()
```
