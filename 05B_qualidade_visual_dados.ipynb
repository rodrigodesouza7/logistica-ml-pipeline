#### Bibliotecas

```python
# Bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
import os
```

#### Caminho do dataset transformado

```python
# Caminho do dataset transformado
file_path = "/content/dataflow_transformado.parquet"

# Leitura do dataset
if os.path.exists(file_path):
    df = pd.read_parquet(file_path)
    print("‚úÖ Dataset carregado com sucesso.")
else:
    print("‚ùå Arquivo n√£o encontrado.")
    df = pd.DataFrame()
```

#### üîé VISUALIZA√á√ïES EXPLORAT√ìRIAS

```python
# -------------------------------------------------------------
# üîé VISUALIZA√á√ïES EXPLORAT√ìRIAS
# -------------------------------------------------------------

if not df.empty:
    # 1. Heatmap de nulos
    print("\nüìå Visualiza√ß√£o de dados ausentes (Missing Values)")
    msno.matrix(df, figsize=(10, 4))
    plt.title("Distribui√ß√£o de Valores Nulos no Dataset")
    plt.show()

    # 2. Distribui√ß√£o de vari√°veis num√©ricas
    print("\nüìå Distribui√ß√£o de Vari√°veis Num√©ricas")
    df.select_dtypes(include=['number']).hist(bins=30, figsize=(12, 6), edgecolor='black')
    plt.tight_layout()
    plt.show()

    # 3. Boxplots por categoria de status
    print("\nüìå Boxplot de Vari√°veis por Categoria de Status")
    features_to_plot = ['delay_minutes', 'cost_usd', 'temp_c']
    for feature in features_to_plot:
        plt.figure(figsize=(8, 4))
        sns.boxplot(x='status', y=feature, data=df)
        plt.title(f"Boxplot de {feature} por Status")
        plt.xlabel("Status")
        plt.ylabel(feature)
        plt.tight_layout()
        plt.show()

    print("\n‚úÖ Relat√≥rios visuais gerados com sucesso.")
else:
    print("‚ö†Ô∏è Dataset vazio. N√£o foi poss√≠vel gerar os relat√≥rios visuais.")
```
